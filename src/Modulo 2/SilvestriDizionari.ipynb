```python
dict = {}
dict = {'nome': 'Mario', 'cognome': 'Rossi', 'eta': 30}

print(dict['nome'])
print(dict.get('cognome'))
print(dict['eta'])
print(dict.get('telefono', 'Non disponibile'))

dict['telefono'] = 348-819-3361

del dict['eta']

valore = dict.pop('cognome')
print(valore)

for key in dict.keys():
    print(key, dict[key])

for value in dict.values():
    print(value)

for key, value in dict.items():
    print(key, value)

if 'nome' in dict:
    print("La chiave 'nome' esiste nel dizionario.")
if 'cellulare' in dict:
    print("La chiave 'cellulare' esiste nel dizionario.")

info_extra = {'hobby': 'calcio', 'citta': 'Roma', 'nome': 'Pietro'}
dict.update(info_extra)
print(dict)
```

```python
#Set
set_vuoto = set()
set_popolato = {1, 2, 3, 4, 5, True, 'Ciao'}
print(set_popolato)
#Aggiunta e rimozione di elementi in un set
colori = {'rosso', 'verde', 'blu'}
colori.add('giallo')
colori.remove('verde')
colori.discard('viola')
print(colori)
#Unione di set
set_a = {1, 2, 3}
set_b = {3, 4, 5}
set_unione = set_a.union(set_b)
print(set_unione)
#Intersezione di set
set_intersezione = set_a.intersection(set_b)
print(set_intersezione)
#Differenza simmetrica di set
set_diff_simmetrica = set_a.symmetric_difference(set_b)
print(set_diff_simmetrica)
#Sottoinsieme e superinsieme
set_c = {1, 2}
set_d = {1, 2, 3, 4, 5}
print(set_c.issubset(set_d))
#Duplicati in una lista usando set
lista_con_duplicati = [1, 2, 2, 3, 4, 4, 5]
set_senza_duplicati = set(lista_con_duplicati)
print(set_senza_duplicati)
```
