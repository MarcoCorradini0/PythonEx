```python
class Conto:
    def __init__(self, nome, conto):
        self.nome = nome
        self.conto = conto

class ContoCorrente(Conto):
    def __init__(self, nome, conto, importo):
        super().__init__(nome, conto)
        self.__saldo = importo

    def preleva(self, importo):
        if importo <= self.__saldo:
            self.__saldo -= importo
        else:
            print("Saldo non sufficiente")

    def deposita(self, importo):
        self.__saldo += importo

    def descrizione(self):
        print("Nome:", self.nome)
        print("Conto:", self.conto)
        print("Saldo:", self.__saldo)

    @property
    def saldo(self):
        return self.__saldo

    @saldo.setter
    def saldo(self, nuovo_saldo):
        self.preleva(self.__saldo)
        self.deposita(nuovo_saldo)
conto1 = ContoCorrente("Mario Rossi", "123456789", 1000)
conto1.preleva(500)
conto1.deposita(200)
conto1.descrizione()
```
